```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('P7.csv')
df
```

```python
df_c = df.dropna()
df_c
```

```python
# creating a day column which stores date and time in datetime fomat
df_c['day'] =pd.to_datetime(df_c['Create Date/Time'],dayfirst=True)
df_c['day'].head()
```

```python
# converting to datetime format to hold yearfirst value
# df['prop_date']=pd.to_datetime(df['day']).pd.strftime('%m/%d/%Y')
df_c['day']=pd.to_datetime(df_c['day'],yearfirst=True)
df_c.head()
```

```python
# getting the time from the date time we've been given 
new_df = pd.read_csv('P7.csv')
new_df.dropna(inplace=True)
df_c['Create Date/Time']=new_df['Create Date/Time']
df_c['Time']=pd.to_datetime(df_c['Create Date/Time']).dt.time
df_c.head()
```

```python
# reading date,month and year values from the date and storing in Prop_date column
df_c['Prop_date']=(df_c['day']).dt.strftime('%d/%m/%Y')
df_c.head()
```

```python
# getting month of the crime from data
month = pd.to_datetime(df_c['Prop_date'],dayfirst=True).dt.strftime('%m')
month
```

```python
# creating a column to store the month and storing the details
df_c['month']=month
df_c.head()
```

```python
import seaborn as sns
```

```python
plt.style.use('seaborn')
top_5_crimes = x3['Event Type'].value_counts().sort_values(ascending=False).head()
temp = x3.groupby('Event Type', as_index=False).agg({"month": "count"})
temp = temp.sort_values(by=['month'], ascending=False).head()
temp = temp.sort_values(by='month', ascending=True)
sns.barplot(x='month', y='Event Type', data=temp, palette="mako")
plt.title("Top 5 Crimes in Lucknow", fontdict = {'fontsize': 30, 'fontname':'Arial', 'color': '#000000'})
plt.xlabel("\nCOUNT OF CRIMES", fontdict = {'fontsize': 15})
plt.ylabel("")
plt.xticks(rotation=90)
plt.show()
```

```python
def hour(x):
    return x.strftime("%H")
x3['Hour_Day'] = x3['Time'].apply(hour)
```

```python
test = x3.sort_values(by=['Hour_Day'])
test['Hour_Day']
```

```python
plt.style.use('seaborn-dark')
sns.set_context('paper')
fig, ax = plt.subplots(figsize=(10, 5))
sns.countplot(x='Hour_Day', data=test, palette="rocket")
plt.title("Crimes in Lucknow based on time", fontdict={'fontsize': 40, 'color': '#bb0e14','fontname':'Agency FB'}, weight="bold")
plt.xlabel("\nHour in the Day", fontdict={'fontsize': 15}, weight='bold')
plt.ylabel("Number of Crimes\n", fontdict={'fontsize': 15}, weight="bold")
plt.show()
```

```python
dispute_df = x3[x3['Event Type']=='Dispute']
hours = [int(x) for x in list(dispute_df['Hour_Day'].unique())]
hours = sorted(hours)
dis = list(dispute_df['Hour_Day'].value_counts().sort_index())
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(x=hours, y=dis, palette='inferno')
plt.title("Disputes over a day", fontdict={'fontsize': 40, 'color': '#bb0e14','fontname':'Agency FB'}, weight="bold")
plt.xlabel("\nHour in the Day", fontdict={'fontsize': 15}, weight='bold')
plt.ylabel("Number of Crimes\n", fontdict={'fontsize': 15}, weight="bold")
plt.show()
```
