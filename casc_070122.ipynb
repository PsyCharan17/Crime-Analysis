```python
import pandas as pd
df = pd.read_csv('P7.csv')
df.dropna(inplace=True)
# df.tail()
```

```python
# creating a day column which stores date and time in datetime fomat
df['day'] =pd.to_datetime(df['Create Date/Time'],dayfirst=True)
df['day'].head()
```

```python
# converting to datetime format to hold yearfirst value
# df['prop_date']=pd.to_datetime(df['day']).pd.strftime('%m/%d/%Y')
df['day']=pd.to_datetime(df['day'],yearfirst=True)

df.head()
```

```python
# getting the time from the date time we've been given 
new_df = pd.read_csv('P7.csv')
new_df.dropna(inplace=True)

df['Create Date/Time']=new_df['Create Date/Time']
df['Time']=pd.to_datetime(df['Create Date/Time']).dt.time
df.head()
```

```python
# reading date,month and year values from the date and storing in Prop_date column
df['Prop_date']=(df['day']).dt.strftime('%d/%m/%Y')
df.head()
```

```python
# getting month of the crime from data
month = pd.to_datetime(df['Prop_date'],dayfirst=True).dt.strftime('%m')
month
```

```python
# creating a column to store the month and storing the details
df['month']=month
df.head()
# df.tail()
```
